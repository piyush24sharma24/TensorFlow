**Matrix multiplication**

In machine learning , matrix multiplication is one of the most common operations.

There are two rules our tensors need to fulfill if we're going to matrix multiply them:

1. The inner dimensions must match
2. The resulting  matrix has the shape of the inner dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix multiplication with Python operator '@'
tensor @ tensor
```

```python
# Create a tensor (3, 2 ) tensor
X = tf.constant([[1, 2],
                 [3, 4],
                 [4, 5]])

y = tf.constant([[7, 4],
                 [8, 4],
                 [6, 9]])
X, y
```

```python
# Let's cahnge the shapr of y
reshape_y= tf.reshape(y, shape=(2, 3))
tf.matmul(X, reshape_y)
```

```python
# Can do the same with transpose
X, tf.transpose(X), tf.reshape(X, shape=(2,3))
```

** The dot product**

Matrix multiplication is also refferd to the dot product.

You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`


```python
X, y
```

```python
# Perform the dot product an X and y (requires X or y to be transposed)
tf.tensordot(tf.transpose(X), y, axes=1)
```

```python
# Perform matrix multiplication between X and y ( transpose)
tf.matmul(X, tf.transpose(y))
```

```python
# Perform matrix multiplication between X and y (reshaped)
tf.matmul(X, tf.reshape(y, shape=(2,3)))
```

```python
# Check the value of y, reshape y and transpose y
print("Normal y:")
print(y, "\n") # "\n" is for new line

print("y reshaped to (2, 3):")
print(tf.reshape(y, (2, 3)),"\n")

print("y transposed:")
print(tf.transpose(y))
```

Generally, when performing matrix multiplication on two different tensors and one of the axes doesn't line up , you will transpose (rather than reshape) to specify the matrix multiplication rules

### Find the positional maximum and minimum



```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check the equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Index on which our minimum value position
F[tf.argmin(F)]
```
